# Train test split

- most important jobs as a data scientist: 
    1. define test set (to be as representative as possible of production data)
    2. select right metric to evaluate algorithm
- everything else will work itself out as long as you get those two right!

## Wrapping it in a function:

```python
# For illustration only. Sklearn has train_test_split()
def split_train_test(df, test_ratio):
    shuffled_indices = np.random.permutation(len(df))
    test_set_size = int(len(df) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return df.iloc[train_indices], df.iloc[test_indices]
```

```python
train_df, test_df = split_train_test(housing_df, 0.2)

len(train_df)
print(", ")
len(test_df)

train_df.to_csv('train.csv')
test_df.to_csv('test.csv')
```

## fancy hash splitter

- If you want to make sure you get the same data in test set every time you run split, but new data may be added over time, split on the hash of an identifying column
- If there is no unique id in your columns, you can either:
    - use row number as id (but this may not be stable if you don't strictly only append data)
    - use some other stable feature (e.g. geographic information like latitude, longitude)
    
zlib.crc32(data[, value])
Computes a CRC (Cyclic Redundancy Check) checksum of data. The result is an unsigned 32-bit integer. 
Changed in version 3.0: Always returns an unsigned value. To generate the same numeric value across all 
Python versions and platforms, use crc32(data) & 0xffffffff.

max value is 2**32

```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    in_test_set = crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32
    return in_test_set # returns boolean

def split_train_test_by_id(df, idxs, test_ratio):
    test_bools = idxs.apply(lambda x: test_set_check(x, test_ratio)) # series of booleans
    train_bools = ~test_bools # inverse of test_set
    return df.loc[train_bools], df.loc[test_bools]
```

```python
crc32(np.int64(1))
", "
crc32(np.int64(2))
", "
crc32(np.int64(3))
```

```python
int(0xffffffff)
2 ** 32
3 **33 
3 **33 & 0xffffffff # applies mask to only use latter eight digits of hexadecimal
```

## Custom transformers

- Need some basic understanding of Python Object Oriented Programming (OOP) for this
- Standard sklearn transformers return numpy arrays, which is kind of annoying, so we write them to work with pandas dataframes instead
- There is a package sklearn_pandas that also makes this easier

Let's create a custom transformer to add extra attributes:

BaseEstimator adds get_params() and set_params() methods
TransformerMixin adds a fit_transform() method

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

from sklearn.base import BaseEstimator, TransformerMixin

class DFImputer(BaseEstimator, TransformerMixin):
    def __init__(self, strategy="median"):
        self.imputer = SimpleImputer(strategy=strategy)
        self.strategy = strategy
        self.fitted = False
    
    def fit(self, X, y=None):
        self._feature_names = X.columns
        self.imputer.fit(X)
        self.fitted = True
        return self
    
    def transform(self, X):
        assert self.fitted, "Need to cal .fit(X) function first!"
        return pd.DataFrame(self.imputer.transform(X[self._feature_names]), 
                            columns=self._feature_names, index=X.index)
    
    def get_feature_names(self):
        return self._feature_names

    
class DFStandardScaler(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.scaler = StandardScaler()
        self.fitted = False
    
    def fit(self, X, y=None):
        if isinstance(X, pd.Series):
            X = pd.DataFrame(X.copy(), columns = [X.name])
        self._feature_names = X.columns.tolist()
        self.scaler.fit(X)
        self.fitted = True
        return self
    
    def transform(self, X):
        assert self.fitted
        if isinstance(X, pd.Series):
            X = pd.DataFrame(X.copy(), columns = [X.name])
        return pd.DataFrame(self.scaler.transform(X[self._feature_names]), 
                            columns=self._feature_names, index=X.index)
    
    def get_feature_names(self):
        return self._feature_names
    
    
class DFCombinedColsAdder(BaseEstimator, TransformerMixin):
    def __init__(self, return_all_columns=True): 
        self.return_all_columns = return_all_columns
    
    def fit(self, X, y=None):
        cols = X.columns.tolist()
        if self.return_all_columns:
            self._feature_names = cols + ['rooms_per_household', 'population_per_household', 'bedrooms_per_room']
        else:
            self._feature_names = ['rooms_per_household', 'population_per_household', 'bedrooms_per_room']
        return self  
    
    def transform(self, X):
        X = X.copy()
        X.loc[:, "rooms_per_household"] = X["total_rooms"] / X["households"]
        X.loc[:, "population_per_household"] = X["population"] / X["households"]
        X.loc[:, "bedrooms_per_room"] = X["total_bedrooms"] / X["total_rooms"]
        return X[self._feature_names]
    
    def get_feature_names(self):
        return self._feature_names
```

```python
def pipeline_get_feature_names(self):
    if hasattr(self, '_feature_names'):
        return self._feature_names
    try:
        self._feature_names = self.steps[-1][1].get_feature_names()
    except Exception as e:
        print(e)
        raise AttributeError("failed to generate feature names for pipeline")
    else:
        return self._feature_names

Pipeline.get_feature_names = pipeline_get_feature_names
```

### categorical values transformer:

```python
class DFOneHotEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, name=None, topx=None, min_samples=None, 
                 drop_na=False, drop_other=False, 
                 add_prefix=False, verbose=0):
        """
        One hot a single column. Adds a column _na for missing values and codes any label not
        seen in the training data as _OTHER. Also makes sure all columns in the
        training data will get created in the transformed dataframe, even if there are categories 
        missing in test or production data.
        
        If topx is given and int only encodes the topx most frequent labels
        if topx is given and float only encodes the labels that cover topx fraction the data
        
        if min_samples is given and int only encodes labels with at least min_samples
        if min_samples is given and float only encodes labels that are at least min_samples fraction of data
        
        -drop_na: drops the missing category
        -drop_other drops the other labels category
        -if name is given encodes columns as  name_label
        """
        self.verbose = verbose
        self.topx = topx
        self.min_samples = min_samples
        self.name = name
        self.drop_na, self.drop_other = drop_na, drop_other
        self.add_prefix = add_prefix

    def fit(self, X, y=None):
        
        if isinstance(X, pd.Series):
            X = X.copy()
        elif isinstance(X, pd.DataFrame) and len(X.columns)==1:
            X = X.iloc[:, 0].copy()
        else:
            raise ValueError("expecting either pd.Series or pd.DataFrame with single column!")
        
        if self.name is None: self.name = X.name
        if self.verbose:
            print('Fit: One-hot coding categorical variable {}...'\
                        .format(self.name))
            
        if self.topx is not None:
            # only take the topx most frequent categories
            if isinstance(self.topx, int):
                self.categories = X.value_counts()\
                                         .sort_values(ascending=False)\
                                         .head(self.topx).index.tolist()
            elif isinstance(self.topx, float):
                top_n_rows = len(X) - (self.topx * len(X))
                self.categories = X.value_counts()\
                                        .sort_values()\
                                        .cumsum()\
                                        .loc[lambda x: x >= top_n_rows].index.tolist()
        elif self.min_samples is not None:
            if isinstance(self.min_samples, int):
                self.categories = X.value_counts()\
                                        .sort_values(ascending=False)\
                                        .loc[lambda x: x >= self.min_samples].index.tolist()
            if isinstance(self.min_samples, float):
                assert self.min_samples >= 0.0 and self.min_samples <= 1.0, \
                    "min_samples should be either int or between 0.0 and 1.0!"
                self.categories = X.value_counts()\
                                        .sort_values(ascending=False)\
                                        .divide(len(X))\
                                        .loc[lambda x: x >= self.min_samples].index.tolist()
                
        else:
            self.categories = X.unique()
            
        # set all the other categories to np.nan
        X.loc[~X.isin(self.categories)] = "OTHER"
        
        onehot_cols = pd.get_dummies(X,
                                         prefix = self.name if self.add_prefix else None,
                                         prefix_sep = "_" if self.add_prefix is not None else "",
                                         dummy_na=True).columns.tolist()
        onehot_cols = [str(col) for col in onehot_cols]
        
        if self.add_prefix:
            if self.drop_na and self.name+"_nan" in onehot_cols:
                onehot_cols.remove(self.name+"_nan")
            if self.drop_other and self.name+"_OTHER" in onehot_cols:
                onehot_cols.remove(self.name+"_OTHER")
        else:
            if self.drop_na and "nan" in onehot_cols:
                onehot_cols.remove("nan")
            if self.drop_other and "OTHER" in onehot_cols: 
                onehot_cols.remove("OTHER")
            
        self.columns = onehot_cols
        return self

    def transform(self, X, y=None):
        if self.verbose:
            print("Transform: One-hot coding categorical variable {}...".format(self.name))

        if isinstance(X, pd.Series):
            X = X.copy()
        elif isinstance(X, pd.DataFrame) and len(X.columns)==1:
            X = X.iloc[:, 0].copy()
        else:
            raise ValueError("expecting either pd.Series or pd.DataFrame with single column!")
            
        # set all categories not present during fit() to "OTHER":
        X.loc[(~X.isnull()) & (~X.isin(self.categories))] = "OTHER"

        # onehot encode using pd.get_dummies
        X_onehot = pd.get_dummies(X, 
                                  prefix = self.name if self.add_prefix else None,
                                  prefix_sep = "_" if self.add_prefix is not None else "",
                                  dummy_na=True)

        # add in columns missing in transform() that were present during fit()
        missing_columns = set(self.columns) - set(X_onehot.columns)
        for col in missing_columns:
            X_onehot[col] = 0
        # make sure columns are in the same order
        X_onehot = X_onehot[self.columns]
        assert set(X_onehot.columns.tolist()) == set(self.columns), \
            "{} doest not equal {}".format(X_onehot.columns.tolist(), self.columns)
        # save the column names so that they can be assigned by DataFrameMapper
        self._feature_names = X_onehot.columns
        return X_onehot

    def get_feature_names(self):
        # helper function for sklearn-pandas to assign right column names
        return self._feature_names
```

### Let's fix the ugly prefixes with some more monkey patching:

```python
def get_feature_names(self, drop_names=False):
        """Get feature names from all transformers.
        Returns
        -------
        feature_names : list of strings
            Names of the features produced by transform.
        """
        from sklearn.utils.validation import check_is_fitted
        check_is_fitted(self)
        feature_names = []
        for name, trans, _, _ in self._iter(fitted=True):
            if trans == 'drop':
                continue
            elif trans == 'passthrough':
                raise NotImplementedError(
                    "get_feature_names is not yet supported when using "
                    "a 'passthrough' transformer.")
            elif not hasattr(trans, 'get_feature_names'):
                raise AttributeError("Transformer %s (type %s) does not "
                                     "provide get_feature_names."
                                     % (str(name), type(trans).__name__))
            if drop_names:
                feature_names.extend([f for f in
                                  trans.get_feature_names()])
            else:
                feature_names.extend([name + "__" + f for f in
                                  trans.get_feature_names()])
        return feature_names
```

```python
ColumnTransformer.get_feature_names = get_feature_names
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import expon, reciprocal

# see https://docs.scipy.org/doc/scipy/reference/stats.html
# for `expon()` and `reciprocal()` documentation and more probability distribution functions.

# Note: gamma is ignored when kernel is "linear"
param_distribs = {
        'kernel': ['linear', 'rbf'],
        'C': reciprocal(20, 200000),
        'gamma': expon(scale=1.0),
    }

svm_reg = SVR()
rnd_search = RandomizedSearchCV(svm_reg, param_distributions=param_distribs,
                                n_iter=50, cv=5, scoring='neg_mean_squared_error',
                                verbose=2, random_state=42)
rnd_search.fit(X_train, y_train)
```

The best model achieves the following score (evaluated using 5-fold cross validation):

Now this is much closer to the performance of the `RandomForestRegressor` (but not quite there yet). Let's check the best hyperparameters found:

```python
rnd_search.best_params_
```

This time the search found a good set of hyperparameters for the RBF kernel. Randomized search tends to find better hyperparameters than grid search in the same amount of time.

Let's look at the exponential distribution we used, with `scale=1.0`. Note that some samples are much larger or smaller than 1.0, but when you look at the log of the distribution, you can see that most values are actually concentrated roughly in the range of exp(-2) to exp(+2), which is about 0.1 to 7.4.

```python
expon_distrib = expon(scale=1.)
samples = expon_distrib.rvs(10000, random_state=42)
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Exponential distribution (scale=1.0)")
plt.hist(samples, bins=50)
plt.subplot(122)
plt.title("Log of this distribution")
plt.hist(np.log(samples), bins=50)
plt.show()
```

The distribution we used for `C` looks quite different: the scale of the samples is picked from a uniform distribution within a given range, which is why the right graph, which represents the log of the samples, looks roughly constant. This distribution is useful when you don't have a clue of what the target scale is:

```python
reciprocal_distrib = reciprocal(20, 200000)
samples = reciprocal_distrib.rvs(10000, random_state=42)
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Reciprocal distribution (scale=1.0)")
plt.hist(samples, bins=50)
plt.subplot(122)
plt.title("Log of this distribution")
plt.hist(np.log(samples), bins=50)
plt.show()
```

The reciprocal distribution is useful when you have no idea what the scale of the hyperparameter should be (indeed, as you can see on the figure on the right, all scales are equally likely, within the given range), whereas the exponential distribution is best when you know (more or less) what the scale of the hyperparameter should be.

Question: Try adding a transformer in the preparation pipeline to select only the most important attributes.

```python
from sklearn.base import BaseEstimator, TransformerMixin

def indices_of_top_k(arr, k):
    return np.sort(np.argpartition(np.array(arr), -k)[-k:])

class TopFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_importances, k):
        self.feature_importances = feature_importances
        self.k = k
    def fit(self, X, y=None):
        self.feature_indices_ = indices_of_top_k(self.feature_importances, self.k)
        return self
    def transform(self, X):
        return X[:, self.feature_indices_]
```

Note: this feature selector assumes that you have already computed the feature importances somehow (for example using a `RandomForestRegressor`). You may be tempted to compute them directly in the `TopFeatureSelector`'s `fit()` method, however this would likely slow down grid/randomized search since the feature importances would have to be computed for every hyperparameter combination (unless you implement some sort of cache).

Let's define the number of top features we want to keep:

```python
k = 5
```

Now let's look for the indices of the top k features:

```python
top_k_feature_indices = indices_of_top_k(feature_importances, k)
top_k_feature_indices
```

Let's double check that these are indeed the top k features:

```python
sorted(zip(feature_importances, attributes), reverse=True)[:k]
```

Looking good... Now let's create a new pipeline that runs the previously defined preparation pipeline, and adds top k feature selection:

```python
preparation_and_feature_selection_pipeline = Pipeline([
    ('preparation', full_pipeline),
    ('feature_selection', TopFeatureSelector(feature_importances, k))
])
```
